# Logistic Regression in Spark

```python
churn_data = spark.read.csv('churn.csv', header=True, inferSchema=True)
churn_data.toPandas()
```

```python
churn_data.printSchema()
```

```python
churn_data.columns
```

```python
churn_data.describe().toPandas()
```

```python
columnlist = [item[0] for item in churn_data.dtypes if item[1].startswith('string')]
columnlist
```

```python
categorical_data=churn_data.select('Surname','Geography','Gender','HasCRCard','IsActiveMember')
categorical_data.show()
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml import Pipeline
```

```python
churn_data=churn_data.drop('RowNumber', 'CustomerId')
churn_data.show()
```

```python
categoricalColumns = ['Surname','Geography','Gender','HasCrCard','IsActiveMember']
stages=[]
for categoricalCol in categoricalColumns:
    stringIndexer = StringIndexer (inputCol = categoricalCol, outputCol=categoricalCol +'Index')
    encoder=OneHotEncoder(inputCols=[stringIndexer.getOutputCol()], outputCols=[categoricalCol + 'classVec'])
    stages += [stringIndexer, encoder]
    
label_stringIdx = StringIndexer (inputCol = 'Exited', outputCol ='label')
stages=stages+[label_stringIdx]

numericCols=['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
assemblerInputs = [c+'classVec' for c in categoricalColumns]+numericCols
assembler=VectorAssembler (inputCols=assemblerInputs, outputCol='vectorized_features')
stages+=[assembler]
scaler=StandardScaler(inputCol='vectorized_features',outputCol='features')
stages=stages+[scaler]
```

```python
pipeline = Pipeline(stages=stages)
```

```python
pipeline_model = pipeline.fit(churn_data)
```

```python
from pyspark.ml import Pipeline
cols = churn_data.columns
pipeline = Pipeline (stages=stages)
pipelineModel=pipeline.fit (churn_data)
churn_data=pipelineModel.transform(churn_data)
selectedCols = ['label','features']+cols
churn_data=churn_data.select(selectedCols)
churn_data.show()
```

```python
training, test = churn_data.randomSplit([0.7, 0.3], seed=1234)
```

```python
from pyspark.ml.classification import LogisticRegression
logr = LogisticRegression(featuresCol='features', labelCol='label')
```

```python
logrModel = logr.fit(training)
predictions=logrModel.transform(training)
predictions.select('features', 'label', 'prediction', 'rawPrediction', 'probability').toPandas().head(5)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
print('AUC', evaluator.evaluate(predictions))
```

```python
accuracy=predictions.filter(predictions.label == predictions.prediction).count()/float(predictions.count())
print('Training Part Accuracy:',accuracy)
```

# Overfitting Check-Out

```python
logrModel = logr.fit(training)
predictions=logrModel.transform(test)
predictions.select('features', 'label', 'prediction', 'rawPrediction', 'probability').toPandas().head(5)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
print('AUC', evaluator.evaluate(predictions))
```

```python
accuracy=predictions.filter(predictions.label == predictions.prediction).count()/float(predictions.count())
print('Test Part Accuracy:',accuracy)
```

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
param_grid = (ParamGridBuilder()
    .addGrid(logr.regParam, [0, 0.5, 1, 2])
    .addGrid(logr.elasticNetParam, [0, 0.5, 1])
    .addGrid(logr.maxIter, [1,5,10])
    .build())
cv = CrossValidator(estimator=logr, estimatorParamMaps=param_grid, evaluator=evaluator, numFolds=5)
cvModel=cv.fit(training)
```

```python
predictions=cvModel.transform(test)
print('AUC:', evaluator.evaluate(predictions))
```

```python
from pyspark import SQLContext
sqlContext = SQLContext(spark)
```

# Best Model Feature Weights

```python
weights = cvModel.bestModel.coefficients
weights = [(float(w),) for w in weights]
weightsDF = sqlContext.createDataFrame(weights,['Feature Weight'])
weightsDF.toPandas().head(10)
```
